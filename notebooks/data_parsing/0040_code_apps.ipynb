```python
hf_dataset = load_dataset(path="codeparrot/apps", name="competition", split="test")
hf_dataset.set_format(type='pandas')
df = hf_dataset[:]
len(df)
```

```python
df = df[df["solutions"].map(len) > 0]
len(df)
```

```python
# Print out each solution + question.
# Used to create the negative versions.
# Currently done till `index = 29`
index = -1
for i, row in enumerate(df.itertuples()):
    if i != index:
        continue
    id = row.problem_id
    question = row.question
    solution = eval(row.solutions)[0]
    print(question)
    print(f"-------{id}-----")
    print(solution)
    print(json.dumps({id: {"negative_solution": solution, "question": question, "solution": solution}}, indent=4))
    
```
